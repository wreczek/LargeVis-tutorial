## 10000 data points

```python
input_file  = "Examples/MNIST/mnist_vec784D.txt"
output_file = "Examples/MNIST/mnist_10000_784.txt"
label_file  = "Examples/MNIST/mnist_label.txt"
num_points = 10000
num_dims = 784

infile  = open(input_file, "r")
outfile = open(output_file, "w+")
oldlabelfile = open(label_file, "r")
newlabelfile = open("Examples/MNIST/mnist_label_10000.txt", "w+")

counter = 0
outfile.write(str(num_points) + " " + str(num_dims) + "\n")

lines = infile.readlines()
labels = oldlabelfile.readlines()
for line in lines[1:]:
    label = int(labels[counter])
    counter += 1
    outfile.write(line)
    newlabelfile.write(str(label) + '\n')

    if counter == num_points:
        break

infile.close()
outfile.close()
oldlabelfile.close()
newlabelfile.close()
```

# bh-SNE for comparison

```python
with open("Examples/MNIST/mnist_10000_784.txt", "r") as infile:
    lines = infile.readlines()
    final = []
    for line in lines[1:]:
        final.append( [ float(x) for x in line.split() ] )
        

#bhSNE:
X = np.array(final)
print("Barnes-Hut: ")
bh_sne_model = TSNE(n_components=2, verbose=1, \
                    perplexity=40, n_iter=300, method='barnes_hut')
t_start = time.time()
X_embedded_bh = bh_sne_model.fit_transform(X)
t_end = time.time()
print(f"bh-SNE was carried out for {t_end - t_start} seconds.\n")
# W tym przypadku 2.5 minuty
```

### 1. Sprawdzamy wyniki dla rnej liczby prbek (-samples) 

Moemy oscylowa liczb prbek w granicach 1 - 1000?

```python
train_input_file = open('Examples/FMNIST/fmnist_with_labels.txt','r')
train_labels_file = open('Examples/FMNIST/fmnist_10k_labels.txt','w+')
train_output_file = open('Examples/FMNIST/fmnist_10k.txt','w+')

points = 10000
dims = 784

# testujemy rne kategorie [0 - 9]
categories = [9, 1, 0, 7] 

train_lines = train_input_file.readlines()

lines_no = 0
train_output_file.write(str(points) + " " + str(dims) + "\n")

for line in train_lines[1:]: # pierwsza linia to l.pkt i wym.     
    label = int(line[-2]) # ostatni znak to enter, 
                          # przedostatni to label

    if label in categories:
        lines_no += 1
        final = line[:-2] + "\n"
        
        train_output_file.write(final)
        train_labels_file.write(str(label) + '\n')
        if lines_no == points:
            break
    
train_input_file.close()
train_output_file.close()
train_labels_file.close()
```

### Preparing the data

```python
###!!! ZMIENIC !!!
input_file = open("Examples/20ng/20ng_nPCA.csv", "r")
output_file = open("Examples/20ng/20ng_nPCA.txt", "w+")
label_file = open("Examples/20ng/20ng_labels.txt", "w+")

points = 18759
dimensions = 30

output_file.write(str(points) + " " + str(dimensions) + "\n")
lines = input_file.readlines()
for line in lines:
    label = line[-3:]
    label_no = int(label, 10)

    text = line.split(',')
    final = " ".join(text)

    output_file.write(final)
    label_file.write(str(label_no) + "\n")

input_file.close()
output_file.close()
label_file.close()
```

```python
categories = [8, 12, 10, 4]

input_training_file = open('Examples/20ng/20ng_nPCA.txt', "r")
output_training_file = open('Examples/20ng/20ng_nPCA_reduced.txt', "w+")
label_training_file = open('Examples/20ng/20ng_nPCA_reduced_labels.txt', "w+")

points_no = 0
lines = input_training_file.readlines()

for line in lines[1:]:
    label_no = int(line[-3:], 10)

    if(label_no in categories):
        points_no += 1
        
input_training_file.seek(0, 0)

counter = 0
output_training_file.write(str(points) + " " + str(dimensions))

lines = input_training_file.readlines()
for line in lines[1:]:
    label = line[-3:]
    label_no = int(label, 10)

    text = line.split(' ')[:dimensions] 
    final = "\n" + " ".join(text)

    if(label_no in categories):
        counter += 1
        output_training_file.write(final)
        label_training_file.write(str(label_no) + "\n")
        if (counter == points_no):
            break
            
input_training_file.close()
output_training_file.close()
label_training_file.close()
```
